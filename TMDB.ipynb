## For SVOD

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


S = 7

data = {"Season_Epi": [], "Episode_name": []}

for i in range(1, S + 1):
    url = f"https://www.themoviedb.org/tv/39340-2-broke-girls/season/{i}?language=hu-HU"
    
    session = requests.Session()
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36",
        "Accept-Language": "en-US,en;q=0.9"
    }
    
    response = session.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    
    # Find all <a> tags with the correct class
    episode_links = soup.find_all('a', class_="no_click open")
    
    for a_tag in episode_links:
        season = a_tag.get("data-season-number")
        episode = a_tag.get("data-episode-number")
        title = a_tag.text.strip()
        
        data["Season_Epi"].append(f"S{season}E{episode}")
        data["Episode_name"].append(title)

# Convert to DataFrame
Episodes = pd.DataFrame(data)
Episodes.to_excel("TMDB.xlsx",index=False)
```
